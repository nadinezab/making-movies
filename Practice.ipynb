```python
df = pd.read_csv('./zippedData/imdb.title.basics.csv.gz')
```

```python
df2 = df[:20]
```

```python
df2
```

```python
df2 = df2.dropna()
```

```python
genres = df2['genres'].unique()
genres[:5]
```

```python
# split genres
splitgenres = []
for genre in genres:
    splitgenres += genre.split(',')
splitgenres[:5]
```

```python
uniquegenres = set(splitgenres)
uniquegenres
```

```python
for genre in uniquegenres:
    df2[genre] = 0
```

```python
df2
```

```python
df3 = df2.copy()
```

```python
df3['genres'] = df3['genres'].map(lambda x: x.split(','))
```

```python
df3
```

```python
for index, genres in enumerate(df3['genres']):
    for genre in genres:
        df3.loc[index,genre] = 1
```
